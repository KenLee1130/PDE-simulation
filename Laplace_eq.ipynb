```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cartesian_2_plor(x, y):
    return np.sqrt(x**2 + y**2), np.arctan2(y, x)
```

```python
# parameter
U = 1.0
R = 1.0
```

```python
# grid
x = np.linspace(-5, 5, 1000)
y = np.linspace(-3, 3, 1000)
X, Y = np.meshgrid(x, y)

r, theta = cartesian_2_plor(X, Y)

r = np.where(r<R, np.nan, r)

# velocity (polar)
u_r = U * (1 - (R**2)/(r**2)) * np.cos(theta)
u_theta = -U * (1 + (R**2)/(r**2)) * np.sin(theta)

# convert to Cartesian
u_x = u_r * np.cos(theta) - u_theta * np.sin(theta)
u_y = u_r * np.sin(theta) + u_theta * np.cos(theta)

# streamlines
plt.figure(figsize=(8, 8))
plt.streamplot(X, Y, u_x, u_y, density=2)

# Disk
circle = plt.Circle((0,0), R, color='black')
plt.gca().add_patch(circle)

plt.xlim(-5,5)
plt.ylim(-3,3)
plt.gca().set_aspect('equal')
plt.title("Flow around cylinder (Constant x velocity + Dipole velocity field)")
plt.show()
```

```python
Gamma = 4*np.pi*U*R*0.8
```

```python
# velocity (polar)
u_r = U * (1 - (R**2)/(r**2)) * np.cos(theta)
u_theta = -U * (1 + (R**2)/(r**2)) * np.sin(theta) + Gamma/(2*np.pi*R)

# convert to Cartesian
u_x = u_r * np.cos(theta) - u_theta * np.sin(theta)
u_y = u_r * np.sin(theta) + u_theta * np.cos(theta)

# streamlines
plt.figure(figsize=(8, 8))
plt.streamplot(X, Y, u_x, u_y, density=3)
# Disk
circle = plt.Circle((0,0), R, color='black')
plt.gca().add_patch(circle)

plt.xlim(-5,5)
plt.ylim(-3,3)
plt.gca().set_aspect('equal')
plt.title(f"Flow around cylinder (Constant x velocity + Dipole velocity field + Circulation) \nGamma={Gamma}")
plt.show()
```
